```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
# 주소입력 변수 생성
url = 'https://www.naver.com'
# 네이버에 요청
response = requests.get(url)
```

```python
html_data = response.text
```

```python
# 파싱
# html의 형태로 변환

soup = bs(html_data, 'html.parser')
```

```python
# find() : HTML안에서 해당 태그의 첫번째 정보를 출력
# find_all() : HTML 안에서 해당 태그의 모든 정보를 리스트 형식으로 출력
print(soup.find('a'))
```

```python
print(soup.find_all('a'))
```

```python
print(soup.find(id='newsstand'))
```

```python
print(soup.find_all('span', limit=10)[1])
```

```python
# 해당 태그에서 class 값을 기분으로 출력
print(soup.find_all('span', attrs={
    'class'  : 'blind'
}))
```

```python
span_list = soup.find_all('span', attrs={
    'class' : 'blind'
})

span_list
```

```python
# 1. 리스트의 값들을 하나씩 출력
# for i in span_list:
#    print(i)
# span_list[0]


# 2. 리스트의 각각 text들을 하나씩 출력
#for i in span_list:
#    print(i.get_text())


    
# 3. text들을 리스트로 생성
# case 1
data_list = []

for i in span_list:
    data_list.append(i.get_text())

data_list
```

```python
# case 2

data_list = []

for i in range(0,len(span_list),1):
    data = span_list[i].get_text()
    data_list.append(data)

data_list
```

```python
# case 3
# map 함수 : map(함수, 리스트 형태의 데이터)

def change(x):
    #print(x.get_text())
    return x.get_text()

list(map(change, span_list))
```

```python
# case 4
# map(lambda 함수, 리스트 형태의 데이터)

list(map(lambda x : x.get_text() , span_list))
```

```python
# 함수 생성
def func_1(_list):
    result = list(map(lambda x : x.get_text(), _list))
    return result

```
