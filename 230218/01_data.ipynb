## 데이터 프레임의 결합
1. 단순한 행, 열 결합


```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'S1' : [1,2,3,4],
    'S2' : [5,6,7,8]
})
df
```

```python
df2= pd.DataFrame([1,2])
df2
```

## concat()

1. 데이터 프레임을 결합하는 함수
2. 매개변수 axis -> 행을 추가할지 열을 추가할지 지정
3. 다나순하게 데이터 프레임을 결합
4. 매개변수 ignore_index ->  기본 값은 false, 인덱스의 값을 그대로 유지할지 지정

```python
## 단순하게 행을 추가

pd.concat([df, df2], axis=0)
```

```python
## 단순하게 행을 추가, 인덱스를 재지정

## case 1 
pd.concat([df, df2], axis = 0).reset_index(drop=True)

##case2
pd.concat([df,df2],axis = 0, ignore_index=True)
```

```python
## 열을 추가

pd.concat([df, df2], axis = 1)
```

```python
df3 = pd.DataFrame({
    'S1' : [1,2]
})
df3
```

```python
pd.concat([df,df3], axis=0)
```

##merge()
1. 데이터 프레임을 결합하는 함수
2. 특정 조건에 맞춰서 열을 추가
3. 매개변수 on -> 조건
4. 매개변수 how ->  데이터 프레임의 기준점 (left, right, inner, outer)


```python
data = [
    ['전기전자', '005930', '삼성전자', 74400],
    ['화학', '051910', 'LG화학', 896000],
    ['전기전자', '000660', 'SK하이닉스',101500 ]
]
columns=['업종','종목코드','종목명','현재가']

df = pd.DataFrame(data=data,columns=columns)
df
```

```python
data = [
    ['은행',2.92],
    ['보험',0.37],
    ['화학',0.06],
    ['전기전자',-2.43]
]
columns = ['업종','등락률']
df2 =pd.DataFrame(data=data, columns=columns)
df2
```

```python
pd.merge(left=df, right = df2, on='업종', how='left')
```

```python
df3= pd.DataFrame({
    '업종' : ['서비스업'],
    '종목코드' : ['035720'],
    '종목명' : ['카카오'],
    '현재가' : [121500]
})
df3
```

```python
df4 = pd.concat([df,df3],axis=0, ignore_index=True)
df4
```

```python
pd.merge(df4,df2, on='업종', how='left')
```
