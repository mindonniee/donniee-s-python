```python
# 라이브러리 로드
import pandas as pd
```

```python
# 코로나 데이터를 로드
# 상위 폴더로 이동 -> CSV 폴더로 이동 -> CORONA.CSV 파일 지정
corona = pd.read_csv('../CSV/corona.csv')
corona.head()
```

```python
# 컬럼 삭제
# DROP() 함수를 이용하여 컬럼을 삭제
# axis라는 매개변수의 값을 기준으로 행과 열을 지정
# inplace 매개변수 값을 기준으로 기준이되는 데이터를 변경을 할지 지정

corona.drop(['Unnamed: 0'], axis =1)

corona.drop(['Unnamed: 0'], axis =1, inplace=True)
#corona = corona.drop(['Unnamed: 0'], axis =1)
```

```python
corona.head()
```

```python
## 컬럼의 이름 변경
## 컬럼의 목록 출력 방법

corona.columns = ['등록일시', '총 사망자','총 확진자','게시글 번호','기준일','기준시간','수정일시',' 누적 의심자','누적 확진율']
```

```python
corona.head()
```

```python
corona.drop(['게시글 번호','기준일','기준시간'], axis=1)
```

```python
# '게시글 번호','기준일','기준시간 컬럼 삭제

corona.drop(['게시글 번호','기준일','기준시간'], axis=1)
```

```python
corona.drop(['게시글 번호','기준일','기준시간'], axis=1, inplace=True)
```

```python
corona
```

```python
# 데이터 프레임의 정렬을 변경
# 등록일시를 기준으로 오름차순 변경
# sort_values([기준이 되는 컬럼의이름])
# ascending 매개변수의 값이 True(기본값)이면 오름차순, false 면 내림차순 정렬

corona.sort_values(['등록일시'], inplace=True)
```

```python
corona
```

```python
# 인덱스의 값을 초기화
# reset_index()
corona.reset_index(drop=True, inplace=True)
```

```python
# 결측치 조회
corona.info()
```

```python
corona.isna().sum()
```

```python
# 데이터 간의 차이를 출력하는 함수 
#diff(n) : n은 인덱스의 이동 횟수 (n은 기본값이 1)

corona['총 확진자'].diff()
```

```python
corona['일일 확진자'] = corona['총 확진자'].diff()
```

```python
corona.head()
```

```python
corona['총 사망자'].diff()

corona['일일 사망자'] = corona['총 사망자'].diff()
```

```python
corona.head()
```

```python
# 인덱스를 이동
# shift(n) n만큼 인덱스 이동

corona['총 사망자'].shift(1)
```

```python
# 결측치를 다른 값으로 대체
# fillna(n) : 결측치를 n으로 채워준다

# 일일 확진자, 일일 사망자의 결측치를 0으로 대체
corona[['일일 확진자', '일일 사망자']].fillna(0)
```

```python
corona[['일일 확진자', '일일 사망자']] = corona[['일일 확진자', '일일 사망자']].fillna(0)
```

```python
corona.head()
```

```python
# fillna() method 매개변수 -> ffill(앞데이터), bfill(뒤데이터)
```

```python
import platform
```

```python
x = [1,2,3,4]
y1 = [3,6,2,7]
y2 = [1,10,3,1]

## 한글이 꺠지는 현상을 막기 위해 폰트 변경
if platform.system() == 'Darwin': #mac의 경우
    plt.rc('font', family='apple gothic')
else :
    plt.rc('font', family='malgun gothic')



# 선의 색
plt.plot(x, y1, 'r')
# 선의 종류
plt.plot(x, y2, '--')
# X축의 이름
plt.xlabel('class')
plt.ylabel('count')
## title
plt.title('graph')
# 범례 추가
plt.legend(['A','B'])
plt.show()
```

```python
# subplot(행의 수, 열의 수 , 위치) #그래프 쪼개기 #여러 그래프 보여주기

# 그래프 2개 그려보기
# subplot을 이용해서 위치를 지정
plt.subplot(2,1, 1)
plt.plot(x,y1)
plt.subplot(2,1,2)
plt.plot(x,y2)

plt.show()
```
