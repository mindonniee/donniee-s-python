```python
## 예제 문제
```

1. csv 폴더에 있는 sales records.csv 파일 로드
2. 'OrderID'. Total Revenue, Total cost, Total profit 컬럼을 삭제
3. 컬럼의 이름을 변경 [대륙, 국가, 채널, 우선순위, 주문일자, 발송일자, 판매단위, 단가, 원가]
4. 대륙, 국가를 기준으로 오름차순 정렬
5. 인덱스를 초기화(기존의 인덱스는 삭제) -> 1) sale.reset_index(drop=true)  / 2) sales.reset_index().drop(['index'], axis=1)

```python
import pandas as pd

SalesRecords = pd.read_csv('../../CSV/Sales Records.csv')
```

```python
SalesRecords
```

```python
SalesRecords.drop(['Order ID','Total Revenue','Total Cost','Total Profit'], axis=1, inplace=True)
```

```python
SalesRecords.columns = ['대륙', '국가', '타입','채널', '우선순위', '주문일자', '발송일자', '판매단위', '단가', '원가']
```

```python
SalesRecords.head()
```

```python
SalesRecords.sort_values(['대륙','국가'], inplace=True)
```

```python
SalesRecords
```

```python
SalesRecords.reset_index(drop=True, inplace=True)
```

```python
SalesRecords
```

```python
# 총 단가 = (판매단위x단가), 총 원가 = (판매단위x원가), 총 수익 = (총 단가 - 총 원가) or 판매단위x(단가-원가)
```

```python
SalesRecords['총 단가'] = SalesRecords['판매단위'] * SalesRecords['단가']
```

```python
SalesRecords.head()
```

```python
SalesRecords['총 원가'] = SalesRecords['판매단위'] * SalesRecords['원가']
```

```python
SalesRecords.head()
```

```python
SalesRecords['총 이윤'] = SalesRecords['총 단가'] - SalesRecords['총 원가']
SalesRecords['총 이윤'] = SalesRecords['판매단위'] * (SalesRecords['단가'] - SalesRecords['원가'])
```

```python
SalesRecords.head()
```

```python
test = SalesRecords.groupby('대륙').sum()[['총 이윤']]

## 내립차순 정렬
test.sort_values(['총 이윤'], ascending=False).head(1)
```

```python
## 국가별 총 이윤이 높은 상위 10개의 국가를 출력
test2 = SalesRecords.groupby('국가').sum()[['총 이윤']]

# 총 이윤을 기준으로 내림차순 정렬
group_df = test2.sort_values(['총 이윤'], ascending=False)
# 상위 10개
top10 = group_df.head(10)
# 하위 10개
bottom10 = group_df.tail(10)
```
